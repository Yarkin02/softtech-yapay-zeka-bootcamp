```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
breast_cancer = pd.read_csv("breast-cancer-data.csv")
df = breast_cancer.copy()
```

```python
# İlk 5 gözlem
print(df.head())
print("- " * 40)

# Son 5 gözlem
print(df.tail())
print("- " * 40)

# Sütun isimleri
print(df.columns)
print("- " * 40)

# Değişken özellikleri
print(df.info())
print("- " * 40)

# Sayısal değişkenlerin istatistikleri
print(df.describe().T)
```

```python
plt.figure(figsize=(10, 6))

# İlk çizgi: Texture Mean
plt.plot(
    df.index,                     # X ekseni: index
    df["texture_mean"],           # Y ekseni: texture_mean
    color="red",                  # Çizgi rengi
    alpha=0.7,                    # Saydamlık
    label="Texture Mean"          # Efsane etiketi
)

# İkinci çizgi: Radius Mean
plt.plot(
    df.index,                     # X ekseni: index
    df["radius_mean"],            # Y ekseni: radius_mean
    color="blue",                 # Çizgi rengi
    alpha=0.7,                    # Saydamlık
    label="Radius Mean"           # Efsane etiketi
)

# Başlık ve eksen etiketleri
plt.title("Texture Mean ve Radius Mean Karşılaştırması")
plt.xlabel("Index")
plt.ylabel("Mean Değerleri")

# Izgara ve efsane
plt.grid(True)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(6, 2))
plt.scatter(df["radius_mean"],
           df["area_mean"],
           color = "blue",
           s = 15,
           alpha = 0.5,
)
plt.xlabel("Radius mean")
plt.ylabel("Area mean")
plt.title("Radius - Area Mean İlişkisi")
```

```python
import matplotlib.pyplot as plt

# Kötü huylu (Malignant)
m = df[df["diagnosis"] == "M"]
m.reset_index(drop=True, inplace=True)

# İyi huylu (Benign)
b = df[df["diagnosis"] == "B"]
b.reset_index(drop=True, inplace=True)

# Şekil boyutunu ayarla
plt.figure(figsize=(15, 10))

# 1. Grafik: Kötü huylu scatter plot
plt.subplot(2, 3, 1)
plt.scatter(m["radius_mean"], m["texture_mean"], color='red', alpha=0.7, label='Malignant')
plt.title("Kötü Huylu - Radius ve Texture")
plt.xlabel("Radius Mean")
plt.ylabel("Texture Mean")
plt.legend()

# 2. Grafik: Kötü huylu histogram
plt.subplot(2, 3, 2)
plt.hist(m["radius_mean"], bins=20, color='salmon', edgecolor='black', alpha=0.7)
plt.title("Kötü Huylu - Radius Histogramı")
plt.xlabel("Radius Mean")
plt.ylabel("Frekans")

# 3. Grafik: Kötü huylu çizgi grafiği
plt.subplot(2, 3, 3)
plt.plot(m["radius_mean"], color='red', linestyle='--', linewidth=1)
plt.title("Kötü Huylu - Radius Çizgi Grafiği")
plt.xlabel("Index")
plt.ylabel("Radius Mean")

# 4. Grafik: İyi huylu scatter plot
plt.subplot(2, 3, 4)
plt.scatter(b["radius_mean"], b["texture_mean"], color='blue', alpha=0.7, label='Benign')
plt.title("İyi Huylu - Radius ve Texture")
plt.xlabel("Radius Mean")
plt.ylabel("Texture Mean")
plt.legend()

# 5. Grafik: İyi huylu histogram
plt.subplot(2, 3, 5)
plt.hist(b["radius_mean"], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("İyi Huylu - Radius Histogramı")
plt.xlabel("Radius Mean")
plt.ylabel("Frekans")

# 6. Grafik: İyi huylu çizgi grafiği
plt.subplot(2, 3, 6)
plt.plot(b["radius_mean"], color='blue', linestyle='--', linewidth=1)
plt.title("İyi Huylu - Radius Çizgi Grafiği")
plt.xlabel("Index")
plt.ylabel("Radius Mean")

# Şekil düzenini sıkıştır ve göster
plt.tight_layout()
plt.show()
```
