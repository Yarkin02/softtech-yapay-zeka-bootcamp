# Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
iris = pd.read_csv("Iris.csv")
df = iris.copy()
```

```python
# İlk 5 gözlem
print(df.head())
print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -")

# Son 5 gözlem
print(df.tail())
print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -")

# Sütun isimleri
print(df.columns)
print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -")

# Değişken özellikleri
print(df.info())
print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -")

# Sayısal değişkenlerin istatistikleri
print(df.describe().T)
```

## Line Plot - Çizgi Grafiği

```python
plt.figure(figsize=(6, 2))  # Grafik boyutlarını 6 birim genişlik ve 2 birim yükseklik olacak şekilde ayarlar.
plt.plot(
    df["Id"],                  # X eksenine Id sütunundaki değerleri yerleştirir.
    df["SepalLengthCm"],       # Y eksenine SepalLengthCm sütunundaki değerleri yerleştirir.
    color="blue",              # Çizginin rengini mavi olarak ayarlar.
    alpha=0.9,                 # Çizginin opaklığını (saydamlık oranını) %90 olarak belirler.
    label="SepalLengthCm Etiket"  # Grafiğe bir etiket ekler; bu, legend'da görünür.
)
plt.plot(
    df["Id"],
    df["PetalLengthCm"],
    color="yellow",
    alpha=0.9,
    label="PetalLengthCm Etiket"
)
plt.title("Iris Çiçeği Görselleştirme")  # Grafiğin başlığını belirler.
plt.xlabel("Id")  # X ekseni için bir isim/etiket belirler.
plt.ylabel("Sepal ve Petal Length")  # Y ekseni için bir isim/etiket belirler.
plt.grid(True)  # Grafik üzerinde grid çizgilerini görünür yapar.
plt.legend()  # Grafikteki etiketlerin bir açıklamasını (legend) gösterir.
plt.show()  # Grafiği oluşturur ve gösterir.
```

## Scatter Plot - Saçılım Grafiği

```python
import matplotlib.pyplot as plt

df_setosa = df[df["Species"] == 'Iris-setosa']  # Iris-setosa türündeki veriyi ayırır
df_versicolor = df[df["Species"] == 'Iris-versicolor']  # Iris-versicolor türündeki veriyi ayırır
df_virginica = df[df["Species"] == 'Iris-virginica']  # Iris-virginica türündeki veriyi ayırır

plt.figure(figsize=(6, 2))  # Grafik boyutlarını 6 birim genişlik ve 2 birim yükseklik olarak ayarlıyoruz

plt.scatter(
    df_setosa["SepalLengthCm"],  # X eksenine SepalLengthCm sütunundaki değerleri yerleştirir
    df_setosa["PetalLengthCm"],  # Y eksenine PetalLengthCm sütunundaki değerleri yerleştirir
    color="black",               # Iris-setosa türü için renk siyah olarak belirlenir
    s=15,                        # Noktaların boyutunu 15 birim olarak ayarlıyoruz
    label="Iris-setosa"          # Iris-setosa türü için label (etiket) ekleriz
)

plt.scatter(
    df_versicolor["SepalLengthCm"],  # X eksenine SepalLengthCm sütunundaki değerleri yerleştirir
    df_versicolor["PetalLengthCm"],  # Y eksenine PetalLengthCm sütunundaki değerleri yerleştirir
    color="yellow",                  # Iris-versicolor türü için renk sarı olarak belirlenir
    s=15,                            # Noktaların boyutunu 15 birim olarak ayarlıyoruz
    label="Iris-versicolor"          # Iris-versicolor türü için label (etiket) ekleriz
)

plt.scatter(
    df_virginica["SepalLengthCm"],  # X eksenine SepalLengthCm sütunundaki değerleri yerleştirir
    df_virginica["PetalLengthCm"],  # Y eksenine PetalLengthCm sütunundaki değerleri yerleştirir
    color="blue",                   # Iris-virginica türü için renk mavi olarak belirlenir
    s=15,                           # Noktaların boyutunu 15 birim olarak ayarlıyoruz
    label="Iris-virginica"          # Iris-virginica türü için label (etiket) ekleriz
)

plt.title("İris Çiçeği Görselleştirme")  # Grafik başlığını ekleriz
plt.xlabel("SepalLengthCm")  # X ekseni etiketini belirleriz
plt.ylabel("PetalLengthCm")  # Y ekseni etiketini belirleriz
plt.grid(True)  # Grafik üzerinde grid çizgilerini açar
plt.legend()  # Grafik üzerindeki etiketlerin açıklamasını (legend) gösterir
plt.show()  # Grafiği görüntüleriz
```
